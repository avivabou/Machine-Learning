## Impurity Measures

(5 points each - 10 points total)

Impurity is a measure of how often a randomly chosen element from the set would be incorrectly labeled if it was randomly labeled according to the distribution of labels in the subset. Implement the functions `calc_gini` and `calc_entropy`. You are encouraged to test your implementation.

````python
def calc_gini(data):
    """
    Calculate gini impurity measure of a dataset.
 
    Input:
    - data: any dataset where the last column holds the labels.
 
    Returns the gini impurity.    
    """
    gini = 1.0
    ###########################################################################
    # TODO: Implement the function.                                           #
    ###########################################################################
    classifications = np.unique(data[:,data.shape[1]-1], return_counts=True)
    for c in classifications[1]:
        gini -= np.square(c/data.shape[0])
    ###########################################################################
    #                             END OF YOUR CODE                            #
    ###########################################################################
    return gini
````

````python
def calc_entropy(data):
    """
    Calculate the entropy of a dataset.

    Input:
    - data: any dataset where the last column holds the labels.

    Returns the entropy of the dataset.    
    """
    entropy = 0.0
    ###########################################################################
    # TODO: Implement the function.                                           #
    ###########################################################################
    classifications = np.unique(data[:,data.shape[1]-1], return_counts=True)
    for c in classifications[1]:
        temp_val = c/data.shape[0]
        entropy -= temp_val*np.log2(temp_val)
    ###########################################################################
    #                             END OF YOUR CODE                            #
    ###########################################################################
    return entropy
````

## Goodness of Split

(10 Points)

Given a feature the Goodnees of Split measures the reduction in the impurity if we split the data according to the feature.
$$
\Delta\varphi(S, A) = \varphi(S) - \sum_{v\in Values(A)} \frac{|S_v|}{|S|}\varphi(S_v)
$$

In our implementation the goodness_of_split function will return either the Goodness of Split or the Gain Ratio as learned in class. You'll control the return value with the `gain_ratio` parameter. If this parameter will set to False (the default value) it will return the regular Goodness of Split. If it will set to True it will return the Gain Ratio.
$$
GainRatio(S,A)=\frac{InformationGain(S,A)}{SplitInformation(S,A)}
$$
Where:
$$
InformationGain(S,A)=Goodness\ of\ Split\ calculated\ with\ Entropy\ as\ the\ Impurity\ function \\
SplitInformation(S,A)=- \sum_{a\in A} \frac{|S_a|}{|S|}\log\frac{|S_a|}{|S|}
$$
NOTE: you can add more parameters to the function and you can also add more returning variables (The given parameters and the given returning variable should not be touch).

````python
def goodness_of_split(data, feature, impurity_func, gain_ratio=False):
    """
    Calculate the goodness of split of a dataset given a feature and impurity function.

    Input:
    - data: any dataset where the last column holds the labels.
    - feature: the feature index.
    - impurity func: a function that calculates the impurity.
    - gain_ratio: goodness of split or gain ratio flag.

    Returns the goodness of split (or the Gain Ration).  
    """
    ###########################################################################
    # TODO: Implement the function.                                           #
    ###########################################################################

    goodness = 0.0
    information_gain = 0.0
    uniques = np.unique(np.column_stack([data[:,feature]]), return_counts=True)
    for i in range(len(uniques[1])):
        information_gain += (uniques[1][i]/len(data)) * impurity_func(data[data[:, feature] == uniques[0][i]])
    
    goodness = impurity_func(data) - information_gain
    if(gain_ratio):
        split_Info = -1 * np.dot((uniques[1]/len(data)), (np.log2(uniques[1]/len(data))))
        if (split_Info == 0):
            goodness = 0
        else:
            goodness = (impurity_func(data) - information_gain) / split_Info
    ###########################################################################
    #                             END OF YOUR CODE                            #
    ###########################################################################
    return goodness 
````

Use a Python class to construct the decision tree. Your class should support the following functionality:

````python
class DecisionNode:

    # This class will hold everything you require to construct a decision tree.
    # The structure of this class is up to you. However, you need to support basic 
    # functionality as described above. It is highly recommended that you 
    # first read and understand the entire exercise before diving into this class.
    
    def __init__(self, feature, value, instances, features):
        self.feature = feature # column index of criteria being tested
        self.value = value
        self.instances = instances
        self.features = features
        self.labels = {'e': 0, 'p': 0}
        ys = np.unique(self.instances[:,-1], return_counts=True)
        self.labels[ys[0][0]]=ys[1][0]
        self.labels[ys[0][1]]=ys[1][1]
        self.children = {}
        self.depth = 0
        self.hasChildren = False
        self.height = 0
        self.pred = None
        self.nodesAmount = 0
        self.nextFeature = None
        toDelete = [feature]
        if not (feature == None):
            self.features = [i for i in self.features if i != self.feature]
             #Calculating the predict and probabilities
            self.pred = max(self.labels, key=self.labels.get) 
        
    def check_for_chi (self, chi, chi_square_table,df_t):
        if (chi == 1):
            return False
        if (min(self.labels, key=self.labels.get) == 0):
            return True;
        dfs = np.unique(self.instances[:,self.nextFeature], return_counts = True)
        nf = {}
        pf = {}
        for c in dfs[0]:
            nf[c] = 0
            pf[c] = 0
        for row in self.instances:
            c = row[self.nextFeature]
            if (row[-1] == 'e'):
                pf[c] += 1
            else:
                nf[c] += 1
        chi_square = 0
        total = self.labels['e'] + self.labels['p']
        P0 = self.labels['e'] / total
        P1 = self.labels['p'] / total
        for df in dfs[1]:
            E0 = df * P0
            E1 = df * P1
            chi_square += np.square(pf[c] - E0) / E0 +  np.square(nf[c] - E1) / E1
        
        return (chi_square >= chi_square_table[df_t][chi])
        
    def add_child(self, child, classification, root):
        self.children[classification] = child
        self.hasChildren = True
        self.nextFeature = child.feature
        child.instances = self.instances[child.instances[:,child.feature] == classification]
        child.depth = (self.depth + 1)
        root.height = max(root.height,child.depth)
        root.nodesAmount += 1
        
````

````python
def build_tree(data, impurity, gain_ratio=False, chi=1, chi_square_table = [[]], max_depth=1000):
    """
    Build a tree using the given impurity measure and training dataset. 
    You are required to fully grow the tree until all leaves are pure. 

    Input:
    - data: the training dataset.
    - impurity: the chosen impurity measure. Notice that you can send a function
                as an argument in python.
    - gain_ratio: goodness of split or gain ratio flag

    Output: the root node of the tree.
    """
    root = None
    ###########################################################################
    # TODO: Implement the function.                                           #
    ###########################################################################
    queue = []
    
    #Creating the root of the tree
    features=[i for i in np.arange(data.shape[1]-1)]
    root = DecisionNode(None,None,data,features)
    queue.append(root)
    
    #while there are nodes at queue
    while (len(queue) > 0):
        #Take the next node
        splits = {}   
        node = queue.pop(0)  
        #Check depth limit
        if (root.height >= max_depth):
            break

        #Continue with node iff the impurity isnt 0
        if (impurity(node.instances) == 0):
            continue

        #Current data to work with
        instances = node.instances
        features = node.features  
            
        #Find the best feature
        for feature in features:
            calc_goodness  = goodness_of_split(instances, feature, impurity, gain_ratio)
            splits[feature] = calc_goodness
        splits = {key:val for key, val in splits.items() if val != 0}
        if (len(splits) == 0):
            continue
        best_feature = max(splits, key=splits.get) 
        best_value = splits[best_feature]
        node.nextFeature = best_feature
            
        #Create children iff the chi is good
        classifications = np.unique(instances[:,best_feature])
        if (chi == 1) or (node.check_for_chi(chi,chi_square_table,len(classifications[0]))):
            for classification in classifications:
                child = DecisionNode(best_feature, classification, instances, features)
                #Adding the children into the queue iff the child is not None
                node.add_child(child,classification,root)
                queue.append(child)
    ###########################################################################
    #                             END OF YOUR CODE                            #
    ###########################################################################
    return root
````

````python
def predict(node, instance):
    """
    Predict a given instance using the decision tree
 
    Input:
    - root: the root of the decision tree.
    - instance: an row vector from the dataset. Note that the last element 
                of this vector is the label of the instance.
 
    Output: the prediction of the instance.
    """
    pred = None
    ###########################################################################
    # TODO: Implement the function.                                           #
    ###########################################################################
    if (len(node.children) == 0):
        return node.pred
    for key in node.children:
        if (instance[node.nextFeature] == key):
            return predict(node.children[key], instance)
    return node.pred
    ###########################################################################
    #                             END OF YOUR CODE                            #
    ###########################################################################
    return node.pred
````

````python
def calc_accuracy(node, dataset):
    """
    Predict a given dataset using the decision tree
 
    Input:
    - node: a node in the decision tree.
    - dataset: the dataset on which the accuracy is evaluated
 
    Output: the accuracy of the decision tree on the given dataset (%).
    """
    accuracy = 0
    ###########################################################################
    # TODO: Implement the function.                                           #
    ###########################################################################
    for instance in dataset:
        if (predict(node,instance) == instance[-1]):
            accuracy += 1
    accuracy = (accuracy/len(dataset)) * 100
    ###########################################################################
    #                             END OF YOUR CODE                            #
    ###########################################################################
    return accuracy 
````
